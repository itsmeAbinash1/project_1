📊 Property & Pricing Analysis


```python
#Q1: - Average listing price by city

query = """
SELECT city, AVG(price) AS avg_price
FROM listings
GROUP BY city;
"""

df_average_price_city = pd.read_sql(query, conn)
print(df_average_price_city)
```

```python
#Q2:- Average price per square foot by property type

query = """
SELECT property_type, AVG(price / sqft) AS avg_price_per_sqft
FROM listings
GROUP BY property_type;
"""

df_avg_price_sqft = pd.read_sql(query, conn)
print(df_avg_price_sqft)
```

```python
#Q3:- Furnishing status impact on price
query = """
SELECT furnishing_status, AVG(price) AS avg_price
FROM listings
JOIN property_attributes
ON listings.listing_id = property_attributes.listing_id
GROUP BY furnishing_status;
"""

df_furnishing_price = pd.read_sql(query, conn)
print(df_furnishing_price)

df_furnishing_price = pd.read_sql(query, conn)
print(df_furnishing_price)
```

```python
#Q4: - Do properties closer to metro stations command higher prices?

query = """
SELECT metro_distance_km, AVG(price) AS avg_price
FROM listings
JOIN property_attributes
ON listings.listing_id = property_attributes.listing_id
GROUP BY metro_distance_km;
"""

df_metro_price = pd.read_sql(query, conn)
print(df_metro_price)
```

```python
#Q5:- Are rented properties priced differently from non-rented ones?

query = """
SELECT is_rented, AVG(price) AS avg_price
FROM listings
JOIN property_attributes
ON listings.listing_id = property_attributes.listing_id
GROUP BY is_rented;
"""

df_rented_price = pd.read_sql(query, conn)
print(df_rented_price)
```

```python
#Q6: - How do bedrooms and bathrooms affect pricing?

query = """
SELECT bedrooms, bathrooms, price
FROM listings, property_attributes
WHERE listings.listing_id = property_attributes.listing_id;
"""

df_bed_bath = pd.read_sql(query, conn)
print(df_bed_bath)
```

```python
#Q7: - Do properties with parking and power backup sell at higher prices?

query = """
SELECT parking_available, power_backup, price
FROM listings, property_attributes
WHERE listings.listing_id = property_attributes.listing_id;
"""

df_parking_power = pd.read_sql(query, conn)
print(df_parking_power)
```

```python
#Q8: - How does year built influence listing price?

query = """
SELECT year_built, price
FROM listings, property_attributes
WHERE listings.listing_id = property_attributes.listing_id;
"""

df_year_price = pd.read_sql(query, conn)
print(df_year_price)
```

```python
#Q9 : - Which cities have the highest median property prices?

# We calculate average price per city (simple alternative to median)

query = """
SELECT city, AVG(price) AS avg_price
FROM listings
GROUP BY city
ORDER BY avg_price DESC;
"""

df = pd.read_sql(query, conn)

print(df)
```

```python
#Q10:- How are properties distributed across price buckets?

query = """
SELECT 
    CASE 
        WHEN price < 200000 THEN 'Under 200K'
        WHEN price < 500000 THEN '200K - 500K'
        WHEN price < 1000000 THEN '500K - 1M'
        ELSE 'Above 1M'
    END AS price_group,
    COUNT(*) AS total
FROM listings
GROUP BY price_group;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q11:- What is the average days on market by city?

# We are calculating how many days properties take to sell in each city

query = """
SELECT 
    l.city, 
    AVG(s.days_on_market) AS avg_days_on_market
FROM listings l, sales s
WHERE l.listing_id = s.listing_id
GROUP BY l.city;
"""

# Run query and store result
df_avg_days = pd.read_sql(query, conn)

# Display result
print(df_avg_days)
```

```python
#Q 12:- Which property types sell the fastest?

query = """
SELECT property_type, AVG(days_on_market) AS avg_days
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id
GROUP BY property_type;
"""

df_fastest = pd.read_sql(query, conn)

print(df_fastest)
```

```python
#Q13-What percentage of properties are sold above listing price?

# We are checking listing price vs sale price
# If sale_price is higher → it means sold above listing price

query = """
SELECT 
    (SUM(s.sale_price > l.price) * 100.0 / COUNT(*)) AS percentage
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id;
"""

df = pd.read_sql(query, conn)

print(df)
```

```python
#Q14 - What is the sale-to-list price ratio by city?

# We are simply showing city, listing price and sale price

query = """
SELECT 
    l.city,
    AVG(s.sale_price * 1.0 / l.price) AS sale_to_list_ratio
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id
GROUP BY l.city;
"""

df = pd.read_sql(query, conn)
print(df)

```

```python
#Q15: - Which listings took more than 90 days to sell?

# We are checking how long each property took to sell
# and filtering only those that took more than 90 days

query = """
SELECT listing_id, days_on_market
FROM sales
WHERE days_on_market > 90;
"""

df = pd.read_sql(query, conn)

print(df)
```

```python
#Q16:-How does metro distance affect time on market?

# We are checking metro distance and days taken to sell
# to understand if distance affects selling speed

query = """
SELECT 
    metro_distance_km,
    AVG(days_on_market) AS avg_days
FROM property_attributes p
JOIN sales s 
ON p.listing_id = s.listing_id
GROUP BY metro_distance_km
ORDER BY metro_distance_km;
"""

df = pd.read_sql(query, conn)

print(df)


```

```python
#Q17 :- What is the monthly sales trend?

# We are looking at sale dates and sale prices
# to see how sales change over time

query = """
SELECT 
    strftime('%Y-%m', date_sold) AS month,
    COUNT(*) AS total_sales
FROM sales
GROUP BY month
ORDER BY month;
"""

df = pd.read_sql(query, conn)

print(df)
```

```python
#Q18: - Which properties are currently unsold?

# We are selecting properties that are NOT present in sales table
# That means they are still unsold

query = """
SELECT listing_id, city, price
FROM listings
WHERE listing_id NOT IN (SELECT listing_id FROM sales);
"""

df = pd.read_sql(query, conn)

print(df)

```

```python
#19. Which agents have closed the most sales?

# We are checking which agent sold how many properties

query = """
SELECT l.agent_id, COUNT(*) AS total_sales
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id
GROUP BY l.agent_id;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q20 - Top agents by total sales revenue

# We are seeing agent-wise sales and price

query = """
SELECT l.agent_id, SUM(s.sale_price) AS total_revenue
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id
GROUP BY l.agent_id;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q21. Which agents close deals fastest?
# We are checking agent and time taken to sell property

query = """
SELECT l.agent_id, AVG(s.days_on_market) AS avg_days
FROM listings l
JOIN sales s 
ON l.listing_id = s.listing_id
GROUP BY l.agent_id;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q22:-Does experience affect deals closed?

# We are checking agent experience and deals closed

query = """
SELECT a.experience_years, COUNT(*) AS total_deals
FROM agents a
JOIN listings l ON a.agent_id = l.agent_id
JOIN sales s ON l.listing_id = s.listing_id
GROUP BY a.experience_years
ORDER BY a.experience_years;

"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q23: - Do high-rated agents close faster

# We are checking rating and performance

query = """
SELECT a.rating, AVG(s.days_on_market) AS avg_days
FROM agents a
JOIN listings l ON a.agent_id = l.agent_id
JOIN sales s ON l.listing_id = s.listing_id
GROUP BY a.rating
ORDER BY a.rating DESC;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q24: - Average commission by agent

# We are checking commission rates of agents

query = """
SELECT agent_id, AVG(commission_rate) AS avg_commission
FROM agents
GROUP BY agent_id;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q25: - Agents with most active listings

# We are counting how many listings each agent has

query = """
SELECT agent_id, COUNT(*) AS total_listings
FROM listings
GROUP BY agent_id;
"""

df = pd.read_sql(query, conn)
print(df)
```

------------------------------------ Buyer & Financing Behavior ------------------------------------------------------


```python
#Q26: - What percentage of buyers are investors vs end users?

# We are checking how many buyers are investors or end users

query = """
SELECT 
    buyer_type,
    COUNT(*) AS total_buyers,
    ROUND(100.0 * COUNT(*) / (SELECT COUNT(*) FROM buyers), 2) AS percentage
FROM buyers
GROUP BY buyer_type;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q27:- Which cities have the highest loan uptake rate?

# We are checking loan usage and city (simple view)

query = """
SELECT 
    COUNT(*) AS total_buyers,
    SUM(loan_taken) AS loan_buyers,
    (SUM(loan_taken) * 100.0 / COUNT(*)) AS percentage
FROM buyers;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q28: - What is the average loan amount by buyer type?

# We are checking buyer type and loan amount

query = """
SELECT buyer_type, AVG(loan_amount) AS avg_loan
FROM buyers
GROUP BY buyer_type;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q29: - Which payment mode is most commonly used?

# We are checking payment methods used by buyers

query = """
SELECT payment_mode, COUNT(*) AS total
FROM buyers
GROUP BY payment_mode;
"""

df = pd.read_sql(query, conn)
print(df)
```

```python
#Q30 : - Do loan-backed purchases take longer to close?

# We are comparing loan taken with time taken to sell

query = """
SELECT loan_taken, COUNT(*) AS total_buyers
FROM buyers
GROUP BY loan_taken;
"""

df = pd.read_sql(query, conn)
print(df)
```
